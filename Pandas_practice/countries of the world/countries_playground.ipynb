```python
import pandas as pd
```

```python
df = pd.read_csv('countries-of-the-world.csv')
```

```python
# Find which columns need to be dropped:
df.columns
```

```python
# Make a drop-list for the columns
remove_list = [x for x in df.columns if "Unnamed" in x.split(":")]
```

```python
df = df.drop(columns=remove_list)
```

```python
rich_countries = df.loc[df["GDP ($ per capita)"] > 22100.0]
```

```python
rich_countries
```

```python
most_rich_country = rich_countries.loc[rich_countries["GDP ($ per capita)"].idxmax(), "Country"]
most_rich_gdp = rich_countries["GDP ($ per capita)"].max()
print(f"The richest country is {most_rich_country} with a GDP of {most_rich_gdp}")
```
